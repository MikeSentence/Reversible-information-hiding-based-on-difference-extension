```python
import matplotlib.pyplot as plt 
import cv2
import numpy as np
```

```python
# 读取图片
img = (plt.imread('1.png') * 255).astype(np.uint8)
print(img.shape)
plt.imshow(img)
# print(img)
```

```python
# 转换为灰度图并且保存
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
cv2.imwrite('Gnar.jpg',gray)
```

```python
# 读取保存的图片
res = plt.imread('Gnar.jpg')
plt.imshow(res,cmap='gray')
```
